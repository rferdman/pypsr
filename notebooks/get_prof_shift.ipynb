```python
import numpy as np
from psrfft import fftfit
import matplotlib.pyplot as plt
from psrread import read_asc_prof
from psrplot import plot_prof
```

```python
prof_file = '/Users/ferdman/Work/pulsar/1756-2251/profs/add_epochs/jb_roach/1756-2251.55928.add.asc'
template_file = '/Users/ferdman/Work/pulsar/1756-2251/profs/template/jb_roach/1756-2251.template.jb_roach.1400.denoise.rot.asc'
```

```python
# Read in ascii prof and template prof -- don't put this is main function, and instead just pass profile dict
prof = read_asc_prof(prof_file)
t_prof = read_asc_prof(template_file)
prof_rot = prof.copy()
print 'n_bins = ', len(prof['i']), len(t_prof['i'])
```

```python
# Get amplitudes and phases from FFTing the input data profile

# First, FFT the profile.  These are all real values, so can speed things up using rfft function
t_prof_fft = np.fft.rfft(t_prof['i'])

# Now I can get teh amplitude and phase information (be weary, since we should have phase = 0. if amp = 0.)
t_prof_amp = np.abs(t_prof_fft)
t_prof_phase = np.angle(t_prof_fft)
```

```python
n_prof_bins = len(t_prof['i'])
print 'n_prof = ', n_prof_bins
n_fft_bins = len(t_prof_fft)
print 'n_fft = ', n_fft_bins
n_amp_bins = len(t_prof_amp)
print 'n_amp = ', n_amp_bins
n_phase_bins = len(t_prof_phase)
print 'n_phase = ', n_phase_bins
#print prof_phase
```

```python
# Now run fftfit to match the profiles and out the shifts and scales
shift,eshift,snr,esnr,b,errb,ngood = fftfit(prof['i'],t_prof_amp[1:],t_prof_phase[1:])
```

```python
print shift, ' +- ', eshift
print b, ' +- ', errb
```

```python
shift_phase = -shift/n_prof_bins
shift_rad = shift_phase*2.0*np.pi
print shift_phase, shift_rad
```

```python
# Now FFT the data profile to get out amps and phases
prof_fft = np.fft.rfft(prof['i'])
prof_amp = np.abs(prof_fft)
prof_phase = np.angle(prof_fft)
```

```python
# Now shift profile phases by the amount calculated by fftfit

prof_phase_rot = np.fmod(np.array([prof_phase[i] + i*shift_rad for i in np.arange(1, n_prof_bins/2)]), 2.0*np.pi)
prof_phase_rot = np.insert(prof_phase_rot, 0, prof_phase[0])  # insert DC component
prof_phase_rot = np.append(prof_phase_rot, prof_phase[len(prof_phase)-1])  # append Nyquist component
print 'n_prof_phase = ', len(prof_phase)
print 'n_prof_phase_rot = ', len(prof_phase_rot)
print prof_phase[0], prof_phase_rot[0:2]
print prof_phase[len(prof_phase)-1], prof_phase_rot[len(prof_phase)-1]

# Now set up new profile to inverse transform
prof_fft_real= np.empty(n_prof_bins/2, dtype=float)
prof_fft_imag = np.empty_like(prof_fft_real)

prof_fft_rot = np.empty(n_prof_bins/2+1, dtype=complex)

prof_fft_rot.real = prof_amp*np.cos(prof_phase_rot)
prof_fft_rot.imag = prof_amp*np.sin(prof_phase_rot)

#prof_fft_ro

#prof_fft_rot2.real = prof_amp*np.cos(prof_phase)
#prof_fft_rot2.imag = -prof_amp*np.sin(prof_phase)

#prof_fft_rot.real = np.append(prof_fft_real, np.flipud(prof_fft_real))
#prof_fft_rot.imag = np.append(prof_fft_real, np.flipud(-prof_fft_imag))


prof_rot['i'] = np.fft.irfft(prof_fft_rot)
#prof_rot['i'] = prof_rot_tmp
```

```python
print 'Rotating prof at phase ', prof_phase[1]/(2.0*np.pi), ' to match template at phase ', t_prof_phase[1]/(2.0*np.pi)
print prof_phase[1]/(2.0*np.pi) - t_prof_phase[1]/(2.0*np.pi)
```
